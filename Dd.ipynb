```python
# buildozer.spec für Colab anpassen
import os

# Aktuelles Verzeichnis ermitteln
current_dir = !pwd
current_dir = current_dir[0]

# Pfade für SDK und NDK setzen
sdk_path = os.path.join(current_dir, '.buildozer/android/platform/android-sdk')
ndk_path = os.path.join(current_dir, '.buildozer/android/platform/android-ndk-r25b')

# buildozer.spec anpassen
with open('buildozer.spec', 'r') as f:
  content = f.read()

# SDK und NDK Pfade anpassen
if 'android.sdk_path =' not in content:
  content = content.replace('[app]', f'[app]\n\n# Pfad zum Android SDK\nandroid.sdk_path = {sdk_path}')
else:
  content = content.replace('android.sdk_path =', f'android.sdk_path = {sdk_path}')

if 'android.ndk_path =' not in content:
  content = content.replace('[app]', f'[app]\n\n# Pfad zum Android NDK\nandroid.ndk_path = {ndk_path}')
else:
  content = content.replace('android.ndk_path =', f'android.ndk_path = {ndk_path}')

# Speichern der angepassten buildozer.spec
with open('buildozer.spec', 'w') as f:
  f.write(content)

print("buildozer.spec wurde für Colab angepasst.")
print("\nAktualisierte buildozer.spec:")
!cat buildozer.spec
```

```python
# Umgebungsvariablen setzen
import os
os.environ['PATH'] = os.path.join(os.getcwd(), '.buildozer/android/platform/android-sdk/cmdline-tools/latest/bin') + ':' + os.environ['PATH']
os.environ['ANDROID_SDK_ROOT'] = os.path.join(os.getcwd(), '.buildozer/android/platform/android-sdk')
os.environ['ANDROID_NDK_HOME'] = os.path.join(os.getcwd(), '.buildozer/android/platform/android-ndk-r25b')

# Buildozer im Debug-Modus ausführen
print("Starte APK-Build mit Buildozer (dies kann 15-30 Minuten dauern)...")
!buildozer -v android debug

print("\nBuild-Prozess abgeschlossen.")
```
